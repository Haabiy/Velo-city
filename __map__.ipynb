# <font color='Orange'>Bike share map analysis</font> 

```python
import requests
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt 
from pathlib import Path
import os
from time import sleep
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://www.velo-city2022.com/Programme/#r5hv0j4'
driver.get(url)
sleep(2)
```

### <font color='Orange'>Opening ceremony</font> 

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
ttbody = soup.find('div', {'class' : 'card matchthis'})
title = ttbody.find('div', {'class' : 'card-title plenary'})
print(title)
```

```python
for body in soup.find_all('div', {'id' : 'mainbody'}):
    card_title = body.find_all('div', {'class': 'card-title'})[0].text.strip()
    card_body =  body.find_all('div', {'class': 'card-body'})[0].text.strip()
    print(card_title)
    print(card_body)
    #print(body)
```

### <font color='Orange'>Name, Profession, Location</font> 

```python
x = input('Ready?')
body = soup.find('div', {'id' : 'mainbody'})
row = body.find('div', {'class' : 'row'})

def extract():
    file = open('Velo-city.csv', 'w', encoding='utf-8')
    try:
        for c1 in row.find_all('div', {'class' : 'col-12 col-md-3 mb-3'}):
            name = c1.find_all('h6')[0].text.strip()
            job = c1.find_all('em')[0].text.strip()
            location = c1.find_all('p')[1].text.strip()
            ff = csv.writer(file)
            ff.writerow([name, job, location])
            print(name, job, location)
    except Exception:
        pass

# trims empty cells in between
def trim():
    df = pd.read_csv('Velo-city.csv')
    df.to_csv('Velo-city.csv', index=False)
    return df
```

```python
for i in range(1):
    extract()
```
